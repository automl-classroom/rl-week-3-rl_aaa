```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("tmp/sarsa_rs/sarsa_rs/sweep_results.csv", names=["alpha", "gamma", "epsilon", "mean_reward"])

# Plot reward vs each hyperparameter
sns.pairplot(df, x_vars=["alpha", "gamma", "epsilon"], y_vars="mean_reward", kind="reg", height=4)
plt.suptitle("Hyperparameter Effects on SARSA Performance", y=1.02)
plt.tight_layout()
plt.savefig("hyperparam_effects.png")
plt.show()
```
